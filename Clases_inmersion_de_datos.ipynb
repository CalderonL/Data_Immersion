## **3.1 Obtención del valor del m2 en cada barrio**

Resolucion desafio 2, calcule el valor de m2 por barrio

```python
df.sample(3)
```

```python
datos_barrio=df.groupby('Barrio').sum()
#agrupamos todos los datos por barrio y sumamos
datos_barrio['Valor_m2_Barrio']=datos_barrio['Precio_Millon']/datos_barrio['Area']
#gaurdamos en una nuev columna el valor de m2 de cada barrio
datos_barrio
```

## **3.2 Distribuye el valor m2  a cada Tipo dependiendo el Barrio**

```python
m2_barrio=dict(datos_barrio['Valor_m2_Barrio'])
df['Valor_m2_Barrio'] = df['Barrio']
df['Valor_m2_Barrio'] = df['Valor_m2_Barrio'].map(m2_barrio)
df.sample(3)
```

```python
top_barrios=df['Barrio'].value_counts()[:10].index
#top de barrios con mayor cantidad de inmuebles
top_barrios
```

## **3.3 Reseteamos la columna Barrio para tener numeros de indice y no el nombre del Barrio**

```python
datos_barrio
datos_barrio.reset_index(inplace=True)
#reseteo de la columna barrio
datos_barrio
```

## **3.4 Creación de .query en la columna Barrio**

```python
datos_barrio.query('Barrio in @top_barrios')
#busqueda
```
