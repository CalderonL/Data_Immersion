```python
from google.colab import drive #importamos driva a google colab
drive.mount('/content/dive') #montamos el drive (nos pedira correro y contrasena)
import pandas as pd #importa la libreria pandas
import matplotlib.pyplot as plt # importamos la libreria de matplot para poder realizar graficos
import numpy as np
df = pd.read_csv('/content/dive/MyDrive/inmersion_de_datos/inmuebles_bogota.csv')#pandas lee el archivo .csv en la ubicación
df.head()
columnas={'Baños':'Banos','Área':'Area'} #creamos un diccionario remplazando los titulos que generen errores
df = df.rename(columns=columnas)#renombramos colums con columnas
```

```python
############# PRIMER RETO ########################################################

#Promedio de área de todos los inmuebles en los barrios en el dataset. El top 10.
print(df[['Barrio','Area']].groupby('Barrio').mean().sort_values(by=['Area'], ascending=False).head(10).plot(kind='bar', color='red'))
plt.title('Los Barrios con mayor promedio en Colombia')
plt.show()
```

```python
############ SEGUNDO RETO #########################################################

#Encontrar la casa más barata
df['Valor_Numero'] = df['Valor'].str.replace('$','').str.replace('.','').astype(np.int64) #creamos una nueva columna modificando la columna Valor
data_casa = (df.Tipo == 'Casa') #generamos una respuesta buleana donde TRUE donde indica si en la columna Tipo hay Casa
casa_barata = df[data_casa]#almacena en una variable todos los valores de Tipo que sean igual a Casa
min_val_casa=casa_barata['Valor_Numero'].min()#indica el valor minimo en la columna Valor_Numero
#print(casa_barata.sort_values(by=['Valor_Numero']).head(10).iloc[0])#imprime la casa mas barata de acuerdo con el valor mas bajo de la columna Valor_Numero
print(casa_barata[casa_barata['Valor_Numero']==min_val_casa])#imprime la casa mas barata de acuerdo con el valor mas bajo de la columna Valor_Numero
```

```python
########## TERCER RETO ###############################################################

#Encontrar el apartamento más barato
data_apartamento = (df.Tipo == 'Apartamento')
apartamento_barato = df[data_apartamento]
min_val_depa=apartamento_barato['Valor_Numero'].min()
print(apartamento_barato[apartamento_barato['Valor_Numero']==min_val_depa])
```

```python
######### CUARTO RETO #################################################################

#Encontrar la casa con el mayor número de baños
data_casa = (df.Tipo == 'Casa')
casa_barata = df[data_casa]
max_ban_cas=casa_barata['Banos'].max()#indica el valor maximo en la columna Banos
print(casa_barata[casa_barata['Banos']==max_ban_cas])#imprime todas las casas que tengan la mayo cantidad de banos
```

```python
######## QUINTO RETO #################################################################

#Encontrar la casa con el mayor número de habitaciones
data_casa = (df.Tipo == 'Casa')
casa_barata = df[data_casa]
max_hab_cas=casa_barata['Habitaciones'].max()
print(max_hab_cas)
print(casa_barata[casa_barata['Habitaciones'] == max_hab_cas])
```

```python
from google.colab import drive #importamos driva a google colab
drive.mount('/content/dive') #montamos el drive (nos pedira correro y contrasena)
import pandas as pd #importa la libreria pandas
import matplotlib.pyplot as plt # importamos la libreria de matplot para poder realizar graficos
import numpy as np
df = pd.read_csv('/content/dive/MyDrive/inmersion_de_datos/inmuebles_bogota.csv')#pandas lee el archivo .csv en la ubicación
df.head()
columnas={'Baños':'Banos','Área':'Area'} #creamos un diccionario remplazando los titulos que generen errores
df = df.rename(columns=columnas)#renombramos colums con columnas
```

```python
print(df.Valor[0].split())
#.split separa el signo de dolar del numero ya que no tiene nada dentro del
#parentesis toma el espacio como indicador de separaci[on]
print(df.Valor.str.split())  # transformamos todos los datos de Valor a str
print(type(df.Valor.str.split())) #tipo pandas.series
valor=df.Valor.str.split(expand=True)
#expand=True convierte de pandas.series a un dataframe
print(valor)
print(type(valor))
#<class 'pandas.core.frame.DataFrame'>
df['Moneda']=valor[0]
#crea una nueva columna en df y otorga el valor de la columna 0 del dataset valor
df['Precio']=valor[1]
#crea una nueva columna en df y otorga el valor de la columna 1 del dataset valor
df.sample(3)
#retorna 3 datos aleatorios del dataset
df.info()
#muestra informacion de toda las columnas del dataset,
#en este caso Moneda y Precio siguen siendo str.
#no se puede realizar operaciones matimaticas todavia,
#si trtamos de sumar solo se concatenaria la informacion.
#Elimina los puntos de los datos de Precio
df['Precio'] = df['Precio'].str.replace('.','',regex=True)
df[['Precio','Barrio']]
df.info()
df.sample(3)
df['Precio_Millon']= df.Precio.astype('float')/1000000
df.info()
pd.set_option('display.precision',2)
pd.set_option('display.float_format', lambda x: '%.2f' % x)
#ajusto los decimales de la tabla
df.describe()
```

**1.Estudiar mejor el histograma de valores, seleccionar 3 tipos de inmuebles (refinar el grafico: Titulo, aumentar el tamaño de label, colores, conclusiónes de la informacion)**

```python
import matplotlib.pyplot as plt
import seaborn as sns
tipos=['Edificio','Lote','Bodega']
inmuebles_selec=df[df['Tipo'].isin(tipos)]
plt.figure(figsize=(10,6))
grafica=sns.histplot(data=inmuebles_selec, x='Precio_Millon', kde=True, hue='Tipo', palette=['Blue','Red','Yellow'])
grafica.set_title('Distribucion 3 tipos de inmuebles')
grafica.set_xlabel("Precio por millon", fontsize=15, labelpad=20)
#grafica.set_ylabel("Cantidad", fontdict=15)
grafica.set_ylabel("Cantidad", fontsize=15, labelpad=20)
plt.xlim((0,16000))
plt.savefig('/content/dive/MyDrive/inmersion_de_datos/valor_inmuebles_desafio.png', format='png')
plt.show()
```

**Conclusión:**

Como se puede observar en la grafica, la mayor cantidad de Bodegas, Lotes y Edificios se encuentran en un precio de entre 0 a 3000 millones de pesos Colombianos, pero también contamos con opciones más costosas en el área de los edificios llegando a un precio máximo de 16000 millones de pesos colombianos, el motivo del incremento en el precio de los edificios se debe a la ubicación, ya que está en zonas más al Norte de país.

```python
valor_area_barrio=df[['Barrio','Area','Precio_Millon']].groupby('Barrio').sum()
valor_area_barrio['valor_m2'] = valor_area_barrio['Precio_Millon']/valor_area_barrio['Area']
umbral=valor_area_barrio['valor_m2'].median()
valor=valor_area_barrio['valor_m2']
colores = ['red' if valor > umbral  else 'orange' if valor == umbral else 'green' for valor in valor_area_barrio['valor_m2'].head(15)]
grafica=valor_area_barrio['valor_m2'].head(15).plot.barh(color=colores)
plt.title("Costo de m2 segun el Barrio", fontsize=20)
plt.xlabel('Precio por m2', fontsize=15)
plt.ylabel('Barrio', fontsize=15)
plt.savefig('/content/dive/MyDrive/inmersion_de_datos/valor_m2_barrio.png', format='png')
plt.show()
```
